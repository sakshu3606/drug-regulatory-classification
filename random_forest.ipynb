```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv(r"drug_regulatory_classification_dataset.csv")
dataset.head()
```

```python
dataset.columns = dataset.columns.str.strip()
dataset.columns = dataset.columns.str.replace(" ", "_")
dataset.columns = dataset.columns.str.lower()
```

```python
# Remove rows where target is NaN
dataset = dataset.dropna(subset=["target_regulatory_class"])
```

```python
dataset.head()
```

```python
x = dataset.drop("target_regulatory_class", axis=1)
y = dataset["target_regulatory_class"]
```

```python
x.head()
```

```python
y.head()
```

```python
numeric_features = x.select_dtypes(exclude=["object"]).columns
categorical_features = x.select_dtypes(include=["object"]).columns
```

```python
numeric_features
```

```python
categorical_features
```

```python
numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocess_pipeline_rf = ColumnTransformer([
    ("num", numeric_pipeline, numeric_features),
    ("cat", categorical_pipeline, categorical_features)
])
```

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
print("Missing values in X:")
print(X_train.isnull().sum())

print("\nMissing values in y:")
print(y_train.isnull().sum())
```

```python
# Full pipeline
pipeline = Pipeline([
    ("preprocess", preprocess_pipeline_rf),
    ("model", RandomForestClassifier(random_state=42))
])
```

```python
# param_dist = {
#     "model__n_estimators": [100,150,200,300],
#     "model__max_depth": [10,20,None],
#     "model__min_samples_split": [2,5,10],
#     "model__min_samples_leaf": [1,2,4],
#     "model__max_features": ["sqrt","log2"]
# }

param_dist = {
    "model__n_estimators": [200,300,400,500,700],
    "model__max_depth": [None,10,20,30,40,50],
    "model__min_samples_split": [2,5,10,15],
    "model__min_samples_leaf": [1,2,4,6],
    "model__max_features": ["sqrt","log2",0.5],
    "model__bootstrap": [True, False],
    "model__class_weight": [None,"balanced","balanced_subsample"]
}
```

```python
random_search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_dist,
    n_iter=60,                # increased search
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    random_state=42,
    verbose=2
)
```

```python
random_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", random_search.best_params_)
```

```python
best_model = random_search.best_estimator_
```

```python
# Predictions
y_pred = best_model.predict(X_test)
```

```python
print(y_pred)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Select first tree
one_tree = best_model.named_steps["model"].estimators_[0]

# Get feature names
feature_names = best_model.named_steps["preprocess"].get_feature_names_out()

plt.figure(figsize=(20, 10))

plot_tree(
    one_tree,
    feature_names=feature_names,
    class_names=best_model.named_steps["model"].classes_,
    filled=True,
    rounded=True,
    max_depth=3  # limit depth for readability
)

plt.title("Random Forest - First Decision Tree")
plt.show()
```
