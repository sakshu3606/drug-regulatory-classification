# 🧬 PharmAI — Drug Regulatory Classification Platform

### Instructions
1. **Run Cell 1** — installs all packages including `feature-engine` *(required — run every time after kernel restart)*
2. **Run Cell 2** — ⚡ re-saves all `.pkl` files to fix compatibility (run once)
3. **Run Cell 3** — starts the Flask server
4. **Click 🚀 Open PharmAI** — opens the UI

> ⚠️ **If you see HTTP 500**: The most common cause is `feature-engine` not being installed.  
> Always run Cell 1 first after any kernel restart.

```python
# ══ Cell 1 · Install ALL dependencies (run after every kernel restart) ══════
import subprocess, sys

pkgs = [
    'flask',
    'flask-cors',
    'pandas',
    'numpy',
    'scikit-learn',
    'xgboost',
    'joblib',
    'requests',
    'feature-engine',   # ← REQUIRED: used by preprocess_pipeline.pkl (Winsorizer)
]

for pkg in pkgs:
    r = subprocess.run([sys.executable, '-m', 'pip', 'install', pkg, '-q'],
                       capture_output=True, text=True)
    print(f"{'✅' if r.returncode == 0 else '❌'} {pkg}")

# TensorFlow (for Deep Learning ANN model)
try:
    import tensorflow as tf
    print(f'✅ tensorflow {tf.__version__} (already installed)')
except ImportError:
    print('⏳ Installing tensorflow...')
    r = subprocess.run([sys.executable, '-m', 'pip', 'install', 'tensorflow', '-q'],
                       capture_output=True, text=True)
    print('✅ tensorflow installed' if r.returncode == 0 else f'⚠️ tensorflow failed')

# Verify feature-engine loaded correctly
try:
    from feature_engine.outliers import Winsorizer
    print('\n✅ feature-engine Winsorizer import OK')
except ImportError:
    print('\n❌ feature-engine import FAILED — restart kernel and run this cell again')

print('\n✅ Done — now run Cell 2')
```

```python
# ══ Cell 2 · Fix PKL Compatibility (run once) ════════════════════════════════
# Re-saves all .pkl files with joblib protocol=4 to fix STACK_GLOBAL / UnpicklingError

import os, sys, pickle, joblib

def _find_dir():
    if os.path.exists(os.path.join(os.getcwd(), 'app.py')):
        return os.getcwd()
    home = os.path.expanduser('~')
    for root, dirs, files in os.walk(home):
        dirs[:] = [d for d in dirs if not d.startswith('.') and d not in
                   ('AppData','anaconda3','miniconda3','node_modules','__pycache__','.git')]
        if 'app.py' in files and any(f.endswith('.pkl') for f in files):
            return root
    return os.getcwd()

nb_dir = _find_dir()
os.chdir(nb_dir)
print(f'📂 Project: {nb_dir}\n')

pkl_files = sorted([f for f in os.listdir(nb_dir) if f.endswith('.pkl')])
print(f'Found {len(pkl_files)} pkl files — re-saving with Python {sys.version[:6]}...\n')

success, failed = [], []

for fname in pkl_files:
    fpath = os.path.join(nb_dir, fname)
    obj, method = None, None

    # Load: joblib → pickle → pickle+latin1
    for try_fn in [
        lambda p: joblib.load(p),
        lambda p: pickle.load(open(p, 'rb')),
        lambda p: pickle.load(open(p, 'rb'), encoding='latin-1'),
    ]:
        try:
            obj = try_fn(fpath)
            break
        except Exception:
            pass

    if obj is None:
        print(f'  ❌ {fname} — could not load (may need re-training)')
        failed.append(fname)
        continue

    # Re-save with joblib protocol=4
    try:
        joblib.dump(obj, fpath, protocol=4)
        print(f'  ✅ {fname}')
        success.append(fname)
    except Exception as e:
        try:
            with open(fpath, 'wb') as f:
                pickle.dump(obj, f, protocol=4)
            print(f'  ✅ {fname} (pickle fallback)')
            success.append(fname)
        except Exception as e2:
            print(f'  ❌ {fname} — save failed: {e2}')
            failed.append(fname)

print(f'\n✅ Fixed: {len(success)}  |  ❌ Failed: {len(failed)}')
if not failed:
    print('🎉 All pkl files fixed!')
    print('👉 Kernel → Restart Kernel → then run Cell 1 → then run Cell 3')
else:
    print('⚠️  Failed files:', failed)
```

```python
# ══ Cell 3 · Launch Server ═══════════════════════════════════════════════════
import threading, time, logging, os, sys, socket
from IPython.display import display, HTML

PORT = 5050

def find_project_dir():
    if os.path.exists(os.path.join(os.getcwd(), 'app.py')):
        print(f'✅ Found project folder: {os.getcwd()}')
        return os.getcwd()
    for root_dir in [os.path.expanduser('~'), os.path.join(os.path.expanduser('~'), 'DataScience'),
                     os.path.join(os.path.expanduser('~'), 'Documents')]:
        for dirpath, dirnames, filenames in os.walk(root_dir):
            dirnames[:] = [d for d in dirnames if not d.startswith('.') and d not in
                           ('AppData','anaconda3','miniconda3','node_modules','__pycache__','.git')]
            if 'app.py' in filenames and 'logistic_model.pkl' in filenames:
                print(f'✅ Found via search: {dirpath}')
                return dirpath
    return os.getcwd()

nb_dir = find_project_dir()
# ── MANUAL OVERRIDE (uncomment if auto-detect is wrong) ─────────────────────
# nb_dir = r'C:\Users\YourName\DataScience\Projects\drug_regulatory_classification'

print(f'\n📂 Using: {nb_dir}')

critical = ['app.py', 'index.html', 'logistic_model.pkl', 'preprocess_pipeline.pkl']
all_ok = True
for f in critical:
    exists = os.path.exists(os.path.join(nb_dir, f))
    print(f"  {'✅' if exists else '❌'} {f}")
    if not exists: all_ok = False

if not all_ok:
    raise SystemExit('❌ Missing files. Set nb_dir manually above and re-run.')

def port_in_use(p):
    with socket.socket() as s:
        return s.connect_ex(('127.0.0.1', p)) == 0

if port_in_use(PORT):
    print(f'\n⚠️  Port {PORT} busy — server already running or restart kernel first.')
else:
    os.chdir(nb_dir)
    if nb_dir not in sys.path:
        sys.path.insert(0, nb_dir)
    for mod in list(sys.modules):
        if mod == 'app' or mod.startswith('app.'):
            del sys.modules[mod]

    def run_server():
        logging.getLogger('werkzeug').setLevel(logging.ERROR)
        from app import app as fa
        fa.run(port=PORT, debug=False, use_reloader=False, threaded=True)

    threading.Thread(target=run_server, daemon=True).start()
    print('\n⏳ Starting server', end='')
    for _ in range(20):
        time.sleep(0.5)
        if port_in_use(PORT):
            print(' ✅')
            break
        print('.', end='', flush=True)
    else:
        print('\n❌ Server did not start. Check errors above.')

flask_url = f'http://127.0.0.1:{PORT}'
display(HTML(f'''
<div style="font-family:system-ui,sans-serif;padding:22px 26px;
            background:linear-gradient(135deg,#03050d,#070d1a);
            border:1px solid #162034;border-radius:16px;
            color:#dde6f0;margin:12px 0;">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:16px;">
    <div style="width:40px;height:40px;border-radius:11px;
                background:linear-gradient(135deg,#00e5ff,#a855f7);
                display:grid;place-items:center;font-size:20px;
                box-shadow:0 0 20px rgba(0,229,255,.3);">⚕</div>
    <div>
      <div style="font-weight:800;font-size:1.1rem;
                  background:linear-gradient(90deg,#00e5ff,#7dd3fc);
                  -webkit-background-clip:text;-webkit-text-fill-color:transparent;">
        PharmAI Server Running
      </div>
      <div style="font-size:.7rem;color:#4d6480;margin-top:2px;">Port {PORT} · CORS enabled</div>
    </div>
  </div>
  <div style="display:flex;gap:10px;flex-wrap:wrap;margin-bottom:14px;">
    <a href="{flask_url}" target="_blank"
       style="padding:10px 22px;border-radius:9px;font-weight:700;font-size:.88rem;
              background:linear-gradient(135deg,#00e5ff,#0ea5e9);
              color:#030710;text-decoration:none;">🚀 Open PharmAI UI</a>
    <a href="{flask_url}/models/status" target="_blank"
       style="padding:10px 22px;border-radius:9px;font-weight:600;font-size:.88rem;
              background:transparent;border:1px solid #1d2e4a;
              color:#00e5ff;text-decoration:none;">🔍 Model Status</a>
  </div>
  <div style="background:#070d1a;border:1px solid #162034;border-radius:8px;
              padding:10px 14px;font-size:.7rem;color:#4d6480;line-height:1.7;">
    📂 <code style="color:#00e5ff">{nb_dir}</code><br>
    If UI shows <b style="color:#ff4757">Server Offline</b> — wait 3s and refresh.
  </div>
</div>
'''))
```
